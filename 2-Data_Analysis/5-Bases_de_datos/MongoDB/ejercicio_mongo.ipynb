```python
url = "mongodb+srv://admin:admin@cluster0.jpwo6w9.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"
```

```python
with open('restaurants.json','r') as r:
    data = json.load(r)


# url = f'{connection_type}+{user}+':'+{password}+{adress}+{resto}'
myclient = pymongo.MongoClient(url)

mydb = myclient["Cluster0"] # INSTANCIAR BASE DE DATOS
restaurantes = mydb['restaurantes']
restaurantes.insert_many(data)
```

```python
# Mostrar todos los documentos de la coleccin restaurantes
list(restaurantes.find())
```

```python
#Mostrar los campos restaurant_id, nombre, distrito y cocina, pero excluya el campo _id para todos los documentos de la coleccin restaurantes

list(restaurantes.find({'borough':'Bronx'},{'_id':0,'restaurant_id':1,'name':1,'borough':1,'cuisine':1}))
```

```python
# Mostrar los primeros 5 restaurantes que se encuentran en el distrito Bronx
list(restaurantes.find({'borough':'Bronx'},{'_id':0}).limit(5))
```
