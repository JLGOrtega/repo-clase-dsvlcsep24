```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

```python
drinks = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv",
                     sep=",",
                     header=0,
                     index_col=None)
```

```python
drinks
```

### Step 3. Assign it to a variable called drinks.

### Step 4. Which continent drinks more beer on average?

```python
drinks.groupby("continent")[["beer_servings"]].mean().sort_values("beer_servings", ascending=False).iloc[0]
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks.groupby("continent")["wine_servings"].describe().reset_index()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks.dtypes
```

```python
numeric_cols = [col for col in drinks.columns if drinks.loc[:, col].dtype != object]
numeric_cols
```

```python
drinks.groupby("continent")[numeric_cols].mean().reset_index()
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
drinks.groupby("continent")[numeric_cols].median().reset_index()
```

### Step 8. Print the mean, min and max values for spirit consumption per continent.

```python
drinks.groupby("continent")[["spirit_servings"]].agg(["mean", "min", "max", "median"])
```
